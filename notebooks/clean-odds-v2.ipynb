```python
import pandas as pd
import ast 

df = pd.read_csv('../data/raw/NBAodds2025-11-10.csv')
df.head(5)
```

```python
# --- Parse the JSON-like strings into actual Python dicts ---
def safe_parse(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return None
    return x

df = df.copy()
df['bookmakers'] = df['bookmakers'].apply(safe_parse)

# --- 1) Explode bookmakers ---
df1 = df.explode('bookmakers', ignore_index=False)

# --- 2) Flatten bookmaker dict ---
bk = pd.json_normalize(df1['bookmakers'])
df1 = pd.concat([df1.drop(columns='bookmakers'), bk], axis=1)

# At this point you should have a column literally named "markets"
# (no prefixes, no renaming)

# --- 3) Explode markets ---
df2 = df1.explode('markets', ignore_index=False)

# --- 4) Flatten market dict ---
mkt = pd.json
```
