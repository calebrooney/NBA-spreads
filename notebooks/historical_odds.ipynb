looks good, lets loop it up. Do rest of 2021-22 regular season, starting from 10/23/21, ending 4/10/23. Can go back and get playoff data after getting rest of regular seasons until now and there is usage quota left

```python
from datetime import date, timedelta

start = date(2021, 10, 23)
end   = date(2022, 4, 10)

def load_historical(start, end, playoffs=False):

    all_days = []
    cur = start

    start_year = start.isoformat()[:4]
    end_year = end.isoformat()[:4]

    while cur <= end:
        day_str = cur.isoformat()
        print(f"Processing {day_str}...")

        try:
            csv_path = fetch_csv(day_str, live=False)
            if csv_path is None:
                cur += timedelta(days=1)
                continue

            raw_df = pd.read_csv(csv_path)
            cleaned_df = clean_odds_data(raw_df, live=False)

            if not cleaned_df.empty:
                cleaned_df["fetch_date"] = day_str  # optional but useful
                all_days.append(cleaned_df)

        except Exception as e:
            print(f"FAILED on {day_str}: {e}")

        cur += timedelta(days=1)

    full_df = pd.concat(all_days, ignore_index=True)

    if playoffs:
        full_df.to_csv(f"../data/processed/odds_{end_year}playoffs.csv", index=False)
    else:
        full_df.to_csv(f"../data/processed/odds_{start_year}_{end_year}.csv", index=False)

    print(f"\nFinal dataset: {len(full_df)} rows")
    return full_df
```

```python
# get remaining regular seasons
# 22-23
start22, end23 = date(2022, 10, 18), date(2023, 4, 9)
# 23-24
start23, end24 = date(2023, 10, 24), date(2024, 4, 9)
# 24-25
start24, end25 = date(2024, 10, 22), date(2025, 4, 13)
# 25-26 (up until recently, jan 6)
start25, end26 = date(2025, 10, 21), date(2026, 1, 6)
```

```python
regular22_23 = load_historical(start22, end23)
regular23_24 = load_historical(start23, end24)
regular24_25 = load_historical(start24, end25)
regular25_26 = load_historical(start25, end26)
```
